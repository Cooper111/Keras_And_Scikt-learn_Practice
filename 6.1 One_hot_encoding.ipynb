```python
### 单词级别的one-hot编码简单示例
import numpy as np

samples = ['The cat sat on the mat.', 'The dog ate my homework.']

token_index = {}
for sample in samples:
    for word in sample.split():
        if word not in token_index:
            token_index[word] = len(token_index) + 1 #为每个单词指定唯一的索引，从1开始
            
max_length = 10 #对样本分词，只考虑每个样本前maxlength个单词

results = np.zeros(shape=(len(samples),
                          max_length,
                         max(token_index.values()) + 1))#[样本数， 样本单词数目，单词onehot]

for i, sample in enumerate(samples):#第i个样本
    for j, word in list(enumerate(sample.split()))[:max_length]:#第j个单词
        index = token_index.get(word)
        results[i, j, index] = 1.#第i样本第j单词one-hot编码
```

```python
#这样编码单词从【1】开始，长度为len+1，所以cat是单词第二个是results【2】
results[0][1]#cat
```

```python
#对字符级别的onehot编码
import string

samples = ['The cat sat on the mat.', 'The dog ate my homework.']
characters = string.printable # 所有可打印的ASCII字符
token_index = dict(zip(characters, range(1, len(characters)+1)))#字段映射,从1开始

max_length = 50
results = np.zeros((len(samples), max_length, max(token_index.values()) + 1))
for i, sample in enumerate(samples):
    for j , character in enumerate(sample[:max_length]):
        index = token_index.get(character)
        results[i, j, index] = 1.
```

# 用Keras实现单词级的one-hot编码

```python
#创建一个分词器，设定只考虑钱1k个常见单词
from keras.preprocessing.text import Tokenizer

samples = ['The cat sat on the mat.', 'The dog ate my homework.']

tokenizer = Tokenizer(num_words=1000)
tokenizer.fit_on_texts(samples)#构建单词索引
#将字符串转换为整数索引组成的列表
sequences = tokenizer.texts_to_sequences(samples)
#也可以直接得到onehot二进制表示
one_hot_results = tokenizer.texts_to_matrix(samples, mode='binary')
#找回单词索引
word_index = tokenizer.word_index
print('Found %s unique tokens.' %len(word_index))
```

## 使用散列技巧的单词级one-hot编码

```python
## 使用散列技巧的单词级one-hot编码
samples = ['The cat sat on the mat.', 'The dog ate my homework.']
dimensionality = 1000# 将单词保存为长度为1000的向量，如果单词数量接近1000或超过，name会遇到很多散列冲突，这会降低编码准确性
max_length = 10

results = np.zeros((len(samples), max_length, dimensionality))
for i, sample in enumerate(samples):
    for j, word in list(enumerate(sample.split()))[:max_length]:
        index = abs(hash(word)) % dimensionality #约束在0-1000内，没有为每个单词分配显式索引
        results[i][j][index] = 1
```
